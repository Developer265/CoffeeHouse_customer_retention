### Step 1: Data Cleaning ###

```python
import pandas as pd
```

```python
# uploading and visualizing dataset
data = pd.read_excel('coffee-house-satisfactory-survey.xlsx')
data.head()
```

```python
# exploring the size of the dataset
data.shape
```

```python
data.describe(include='all')
```

```python
# removing duplicates
data = data.drop_duplicates()
```

```python
# exploring the columns
data.info()
```

```python
# checking missing values
data.isna().sum()
```

```python
# create a value to store rows with missing values
mask = data.isna()
mask1 = mask.any(axis=1)
```

```python
missing_rows = data[mask1]
missing_rows
```

```python
data['6. How do you usually enjoy Coffee House?'].value_counts()
```

```python
# Changing missing value
data['6. How do you usually enjoy Coffee House?'] = data['6. How do you usually enjoy Coffee House?'].fillna('never')
```

```python
data['6. How do you usually enjoy Coffee House?'].value_counts()
```

```python
data['19. How do you come to hear of promotions at Coffee House? Check all that apply.'].value_counts()
```

```python
# Changing missing value with the most frequent value
data['19. How do you come to hear of promotions at Coffee House? Check all that apply.'] = data['19. How do you come to hear of promotions at Coffee House? Check all that apply.'].fillna(data['19. How do you come to hear of promotions at Coffee House? Check all that apply.'].mode()[0])
```

```python
data['19. How do you come to hear of promotions at Coffee House? Check all that apply.'].value_counts()
```

```python
# checking there is no more missing values
data.isna().sum()
```

### Step 2: Data Exploration ###

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# exploring data for the customers who are not interested in the CoffeeHouse 
df_coffee_house = data[data['20. Will you continue buying Coffee House?'] == 'No']
df_coffee_house.head()
```

```python
# plotting a bar chart to explore how customers enjoy the place 
sns.countplot(x='6. How do you usually enjoy Coffee House?', data=df_coffee_house, order=df_coffee_house['6. How do you usually enjoy Coffee House?'].value_counts().index)
plt.xlabel('How do you usually enjoy Coffee House')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# plotting a bar chart to explore the quality of the CoffeeHouse
sns.countplot(x='12. How would you rate the quality of Coffee House compared to other brands (Coffee Bean, Old Town White Coffee..) to be:', data=df_coffee_house, order=df_coffee_house['12. How would you rate the quality of Coffee House compared to other brands (Coffee Bean, Old Town White Coffee..) to be:'].value_counts().index)
plt.ylabel('Count')
plt.xlabel('quality')
plt.show()
```

```python
# plotting a bar chart to explore the service feedback for the CoffeeHouse
sns.countplot(x='17. How would you rate the service at Coffee House? (Promptness, friendliness, etc..)', data=df_coffee_house, order=df_coffee_house['17. How would you rate the service at Coffee House? (Promptness, friendliness, etc..)'].value_counts().index)
plt.xlabel('service')
plt.ylabel('Count')
plt.show()
```

```python
# plotting a bar chart to explore how customers find the prices at the CoffeeHouse
sns.countplot(x='13. How would you rate the price range at Coffee House?', data=df_coffee_house, order=df_coffee_house['13. How would you rate the price range at Coffee House?'].value_counts().index)
plt.xlabel('price rate')
plt.ylabel('Count')
plt.show()
```

```python
# plotting a bar chart to explore how much customers are willing to spend in a visit
sns.countplot(x='11. On average, how much would you spend at Coffee House per visit?', data=df_coffee_house, order=df_coffee_house['11. On average, how much would you spend at Coffee House per visit?'].value_counts().index)
plt.xlabel('spendings')
plt.ylabel('Count')
plt.show()
```

```python
# plotting a bar chart to explore the importance of sales and promotions for customers
sns.countplot(x='14. How important are sales and promotions in your purchase decision?', data=df_coffee_house, order=df_coffee_house['14. How important are sales and promotions in your purchase decision?'].value_counts().index)
plt.xlabel('Sales and promotions')
plt.ylabel('Count')
plt.show()
```
